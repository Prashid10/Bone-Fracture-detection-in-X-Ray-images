```python
# split into x and y 
data.columns
```

```python
# mapping the label

enc_label = {'Fractured': 0, 'Non-Fractured': 1}
```

```python
data['label'] = data['label'].map(enc_label)
```

```python
x = data.drop('label', axis = 1)
y = data.label
```
