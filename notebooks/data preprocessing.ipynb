```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import cv2
```

```python
# load dataset
data = pd.read_csv(r'image_labels_paths.csv')
data.head()
```

```python
# Def Image Processing 
def image_processing(img_path):
    img = cv2.imread(img_path)
    # convert to grayscale
    img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    # resize to 100*100
    img_res = cv2.resize(img_gray, (100,100))
    # convert to numpy array
    img_ary = np.array(img_res).flatten()
    # convert to dataFrame
    img_df = pd.DataFrame(img_ary).T
    img_df.head()
    return img_df
```

```python
image_processing(data['Fractured'][0])
```

```python
final_df = pd.DataFrame()
```

```python
# for feature in data.columns: 
#     for img in data[feature]:
#         ret_df = image_processing(img)
#         final_df = pd.concat([final_df, ret_df])

for img in data['Fractured']:
    df_x = image_processing(img)
    final_df = pd.concat([final_df, df_x])
final_df['label'] = 'Fractured'
```

```python
final_df.index = range(len(final_df))
```

```python
final_df.to_csv('bone_fractured_unlabelled.csv', index = False)
```

```python
for img in data['Non-Fractured']:
    df_x = image_processing(img)
    final_dfx = pd.concat([final_dfx, df_x])

final_dfx['label'] = 'Non-Fractured'
```
