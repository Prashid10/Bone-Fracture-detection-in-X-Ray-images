```python
lables = os.listdir(fact_img)
lables
```

```python
# get label path 
label_path = [os.path.join(fact_img, label) for label in lables]
label_path
```

```python
# get dictionary with label names 
label_path_dict = dict.fromkeys(lables)
label_path_dict
```

```python
for label in label_path_dict.keys():
    label_path = os.path.join(fact_img, label)
    img_path = [os.path.join(label_path, img) for img in os.listdir(label_path)]
    # print(img_path)
    label_path_dict[label] = img_path
    # img_path.clear()
```

```python
pd.DataFrame(label_path_dict).to_csv('image_labels_paths.csv', index = False)
```
